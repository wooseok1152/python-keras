```python
from tensorflow.keras.datasets import reuters
from tensorflow.keras.callbacks import EarlyStopping

(train_dataset, train_labels), (test_dataset, test_labels) = reuters.load_data(num_words = 10000)
```

```python
train_dataset
```

```python
word_index = reuters.get_word_index()
len(word_index)
```

```python
reverse_word_index = dict([(value, key) for (key, value) in word_index.items()])
reverse_word_index
```

```python
decoded_newsire = " ".join([reverse_word_index.get(i - 3, "?") for i in train_dataset[0]])
decoded_newsire
```

```python
def vectorize_sequences(dataset, dimension=10000):
    
    results = np.zeros((len(dataset), dimension))
    for i, sequence in enumerate(dataset):
        
        results[i, sequence] = 1
        
    return results

new_train_dataset = vectorize_sequences(train_dataset)
new_test_dataset = vectorize_sequences(test_dataset)
```

```python
from tensorflow.keras.utils import to_categorical

new_train_labels = to_categorical(train_labels)
new_test_labels = to_categorical(test_labels)
```

```python
from tensorflow.keras import models
from tensorflow.keras import layers

model = models.Sequential()
model.add(layers.Dense(64, activation = "relu", input_shape = (10000, )))
model.add(layers.Dense(64, activation = "relu"))
model.add(layers.Dense(46, activation = "softmax"))
```

```python
model.compile(optimizer = "rmsprop", loss = "categorical_crossentropy", metrics = ["accuracy"])
```

```python
val_dataset = new_train_dataset[:1000]
new_train_dataset = new_train_dataset[1000:]
val_labels = new_train_labels[:1000]
new_train_labels = new_train_labels[1000:]
```

```python
#validation_loss가 올라가는 추세를 보이면 조기종료 하도록 설정
early_stopping = EarlyStopping()
```

```python
history = model.fit(new_train_dataset, new_train_labels, epochs = 20, batch_size = 512, validation_data=(val_dataset, val_labels), callbacks=[early_stopping])
```

```python
history.history
```

```python
%matplotlib inline

loss = history.history["loss"]
val_loss = history.history["val_loss"]

epochs = range(1, 10)

plt.plot(epochs, loss, "bo", label = "Training loss")
plt.plot(epochs, val_loss, "b", label = "Validation loss")
plt.title("Training and validation loss")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.legend()

plt.show()
```

```python
accuracy = history.history["accuracy"]
val_accuracy = history.history["val_accuracy"]

epochs = range(1, 10)

plt.plot(epochs, accuracy, "bo", label = "Training accuracy")
plt.plot(epochs, val_accuracy, "b", label = "Validation accuracy")
plt.title("Training and validation accuracy")
plt.xlabel("Epochs")
plt.ylabel("Accuracy")
plt.legend()

plt.show()
```

```python
model.evaluate(new_test_dataset, new_test_labels)
```
